### Dealing with outliers

```python
# Outliers can heavily distort Ordinary Least Squares (OLS) regression results,
# so it's important to address them before model training.

# A common rule of thumb is to trim the top 0.5%–1% of extreme values that may
# represent data entry errors or unusually large observations that bias the model.

# As this data has a strong right skew and would be best to do a log transformation with the outliers
# opposed to trimming them.

cols_to_log = [
    'population',
    'Carbon Dioxide emissions (million metric tons)',
    'Coal',
    'Petroleum',
    'Gas',
    'Nuclear'
]

# Ensure we’re working on a copy of the DataFrame
cleaned_co2_data = cleaned_co2_data.copy()

# Apply the log transformation safely using .loc[row_indexer, col_indexer]
for col in cols_to_log:
    cleaned_co2_data.loc[:, f'{col}_log'] = np.log1p(cleaned_co2_data.loc[:, col])
```

### 1. Check the OLS assumption for linearity


```python

from scipy.stats import pearsonr

# ---------------------------------------------------------------
# Define columns
# ---------------------------------------------------------------
y_col = 'Carbon Dioxide emissions (million metric tons)'
y_log = f'{y_col}_log'

cols_to_log = [
    'population',
    'Coal',
    'Petroleum',
    'Gas',
    'Nuclear'
]
log_cols = [f'{col}_log' for col in cols_to_log]

# ---------------------------------------------------------------
# Helper function to add correlation annotation
# ---------------------------------------------------------------
def annotate_corr(ax, x, y):
    r, p = pearsonr(x, y)
    ax.text(0.05, 0.9, f"r = {r:.2f}", transform=ax.transAxes,
            fontsize=10, color='darkred', fontweight='bold')

# ---------------------------------------------------------------
# Raw X vs log(CO2)
# ---------------------------------------------------------------
n_cols = 3
n_rows = int(len(cols_to_log) / n_cols) + (len(cols_to_log) % n_cols > 0)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(cols_to_log):
    sns.regplot(
        x=cleaned_co2_data[col],
        y=cleaned_co2_data[y_log],
        ax=axes[i],
        scatter_kws={'alpha': 0.6, 'color': 'steelblue'},
        line_kws={'color': 'black'}
    )
    annotate_corr(axes[i], cleaned_co2_data[col], cleaned_co2_data[y_log])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('log(1 + CO2 Emissions)')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('Raw Predictors vs log(CO2 Emissions)\n(with Regression Lines & Correlation Coefficients)', fontsize=15)
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------
# log(X) vs log(CO2)
# ---------------------------------------------------------------
n_cols = 3
n_rows = int(len(log_cols) / n_cols) + (len(log_cols) % n_cols > 0)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(log_cols):
    sns.regplot(
        x=cleaned_co2_data[col],
        y=cleaned_co2_data[y_log],
        ax=axes[i],
        scatter_kws={'alpha': 0.6, 'color': 'darkorange'},
        line_kws={'color': 'black'}
    )
    annotate_corr(axes[i], cleaned_co2_data[col], cleaned_co2_data[y_log])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('log(1 + CO2 Emissions)')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('log(Predictors) vs log(CO2 Emissions)\n(with Regression Lines & Correlation Coefficients)', fontsize=15)
plt.tight_layout()
plt.show()
```

#### Calculate correlations to decide which version of each predictor works best

```python
y = cleaned_co2_data['Carbon Dioxide emissions (million metric tons)_log']

results = []
for col in ['population', 'Coal', 'Petroleum', 'Gas', 'Nuclear']:
    r_raw, _ = pearsonr(cleaned_co2_data[col], y)
    r_log, _ = pearsonr(cleaned_co2_data[f'{col}_log'], y)
    results.append((col, round(r_raw, 3), round(r_log, 3)))

corr_compare = pd.DataFrame(results, columns=['Variable', 'r (raw vs log(Y))', 'r (log(X) vs log(Y))'])
corr_compare
```

r measures linear strength between variables.

It ranges from -1 to +1:

- r ≈ 0 → no linear relationship
- r ≈ 0.5 → moderate linear relationship
- r ≥ 0.7 → strong linear relationship

Based on the finding, the model should have the log of population, coal, petroleum, gas, and CO2 while all other variables can remain as is.

#### 2. Check for Multicollinearity
Multicollinearity is observed when two or more variables have a high correlation.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = linear_co2_data[['population_log',
        'Coal_log','Petroleum_log', 'Gas_log','Biomass', 'Geothermal', 'Solar', 'Wind', 'Hydro',
        'Nuclear', 'Other' ]]

# create a new data frame which will include all the VIFs
vif = pd.DataFrame()

# make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
# include names so it is easier to explore the result
vif["Features"] = variables.columns

vif
```

### 5. Check for Autocorrelation

- The correlation of a variable with itsself over successive time intervals
- Use the Durbin Watson test - 0 - 4. 2 indicates no autocorrelation while values below 1 and above 3 cause for alarm use

```python
# Calculate the Durbin-Watson statistic on the residuals
from statsmodels.stats.stattools import durbin_watson

dw_statistic = durbin_watson(results.resid)
print("Durbin-Watson Statistic:", dw_statistic)
```

- Results 2: Indicates no autocorrelation.
- Results 0 to <2: Indicates positive autocorrelation.
- Results >2 to 4: Indicates negative autocorrelation.

The Results are 1.5756 meaning there is positive autocorrelation but as it is not below 1 then this is not a cause for alarm as this is not time series data.

## Regression Training with Sklearn Model
---

```python
# Drop all irrelevant columns
regression_data = linear_co2_data_scaled[['Carbon Dioxide emissions (million metric tons)_log', 'population_log_scaled','Coal_log_scaled']]
regression_data.head(10)
```

### Declare the inputs and the targets

```python
# The target(s) (y - dependent variable) is 'Carbon Dioxide emissions (million metric tons)_log'
targets = regression_data['Carbon Dioxide emissions (million metric tons)_log']

# The inputs are everything BUT the y - dependent variable
inputs = regression_data.drop(['Carbon Dioxide emissions (million metric tons)_log'],axis=1)
```

### Train Test Split

```python
# Import the module for the split
from sklearn.model_selection import train_test_split

# Split the variables with an 80-20 split and some random state
# To have the same split as mine, use random_state = 365
# first argument refers to the x's & second argument refers to the y's
x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=365)
```

### Create the Regression

```python
# Create a linear regression object
reg = LinearRegression()
# Fit the regression with the TRAIN inputs and targets
reg.fit(x_train,y_train)
```

### 4. Check for Normality (Histogram) and Homoscedasticity (Scatter plot)

```python
# Create a variable to store the predictions based on the training data
y_hat = reg.predict(x_train)
```

### Histogram of the Residuals

```python
# Plot the PDF of the residuals and check for anomalies
sns.displot(y_train - y_hat)

# Include a title
plt.title("Residuals PDF", size=18)
```

For the best case scenario this plot should be normally distributed, which it is so there is normality.

### Scatter plot of the Residuals

```python
# Calculate residuals
residuals = y_train - y_hat   

# Create the scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=y_hat, y=residuals, alpha=0.6, color='steelblue', edgecolor=None)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)

plt.title('Residuals vs Predicted Values (Check for Homoscedasticity)')
plt.xlabel('Predicted Values (ŷ)')
plt.ylabel('Residuals (y - ŷ)')
plt.show()
```

There is a random scatter around 0 so there is homoscedasticity — the assumption that your model’s residuals have constant variance across all levels of the fitted values.

### Scatter Plot of the Targets and Predictions

```python
# The simplest way to compare the targets (y_train) and the predictions (y_hat) is to plot them on a scatter plot
# The closer the points to the 45-degree line, the better the prediction
plt.scatter(y_train, y_hat)
# Let's also name the axes
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)

# Making the y-axis to be the same as to interpret the '45-degree line'

plt.show()
```
